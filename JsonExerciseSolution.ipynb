# Problem 2 - Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
# declare dictionaries to get project code and name , project code and it's count, project name and it's count
# the data has many project names absent but project codes are there and the map btw code & name can be obtained
dict_code_projName = dict()
dict_projCode_nums = dict()
dict_projName_nums = dict()
```

```python
#iterate through 'mjtheme_namecode' column and fill the dictionaries
for value in json_df.mjtheme_namecode.values:
    for dic_item in value:
        code = ''
        name = ''
        for key, val in dic_item.items():
            if key == 'code' and val != '':
                code = val
                dict_projCode_nums[code] = dict_projCode_nums.get(code, 0) + 1
            elif key == 'name' and val != '':
                name = val
        if code != '' and name != '':
            dict_code_projName[code] = name
            
for key, val in dict_projCode_nums.items():
    dict_projName_nums[dict_code_projName[key]] = val
```

```python
#sort the dictionary with values to get most number of projects first
import operator
sorted_list_projName_nums = sorted(dict_projName_nums.items(), key=operator.itemgetter(1), reverse=True)
sorted_list_projName_nums[:10]
```

# Problem 3 - In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
#Create a new data frame by copying from existing
json_df_copy = json_df.copy()
```

```python
#Using dict_code_projName fill all the missing project names with codes
for value in json_df_copy.mjtheme_namecode.values:
    for dic_item in value:
        code = ''
        nameIsThere = True
        for key, val in dic_item.items():
            if key == 'code' and val != '':
                code = val
            elif key == 'name' and val == '':
                nameIsThere = False
        if code != '' and nameIsThere == False:
            dic_item['name'] = dict_code_projName[code]
```
