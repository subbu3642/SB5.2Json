Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

## Solution

importing pandas module

```python
import pandas as pd
```

```python
#read json data frame
json_df = pd.read_json('data/world_bank_projects.json')
json_df
```

# Problem 1 - Find the 10 countries with most projects

```python
json_df['countryshortname'].value_counts()[:10]
```

# Problem 2 - Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
json_df['mjtheme_namecode'].value_counts()[:10]
```

# Problem 3 - In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
json_df.theme_namecode.notnull().value_counts()
```

```python
json_new_df = json_df.copy()
```

```python
json_new_df.theme_namecode = json_new_df.theme_namecode.ffill()
```

```python
json_new_df.theme_namecode.notnull().value_counts()
```
